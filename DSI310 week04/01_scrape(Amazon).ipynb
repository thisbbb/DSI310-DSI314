```python
import bs4
import pandas as pd
import requests
import selenium
from bs4 import BeautifulSoup
from selenium import webdriver 
from selenium.webdriver.firefox.options import Options
import time
```

```python
options = Options()
options.binary_location = r'C:\Program Files\Mozilla Firefox\firefox.exe'
driver = webdriver.Firefox(executable_path=r'C:\Users\COM01\Documents\Pliwlom\DSI 314\geckodriver.exe', options=options)
driver.get('https://www.amazon.com/s?k=plant+based+food&crid=3VE2HS2IT3V3N&sprefix=plant+based+food%2Caps%2C385&ref=nb_sb_noss_2')
```

```python
driver.execute_script("document.body.style.MozTransform='scale(0.1)';")
driver.execute_script('document.body.style.MozTransformOrigin = "0 0";')
```

```python
data = driver.page_source
soup = bs4.BeautifulSoup(data)
time.sleep(6)
```

```python
all_Name = soup.find_all('span',{'class':"a-size-base-plus a-color-base a-text-normal"})
```

```python
all_Name_list = []
for Name in all_Name:
    all_Name_list.append(Name.text)
all_Name_list
```

```python
all_Rating = soup.find_all('span',{'class':"a-icon-alt"})
```

```python
all_Rating_list = []
for Rating in all_Rating:
    all_Rating_list.append(Rating.text)
all_Rating_list
```

```python
all_Review = soup.find_all('span',{'class':"a-size-base puis-light-weight-text s-link-centralized-style"})
```

```python
all_Review_list = []
for Review in all_Review:
    all_Review_list.append(Review.text)
all_Review_list
```

```python
all_Price = soup.find_all('span',{'class':"a-price-whole"})
```

```python
all_Price_list = []
for Price in all_Price:
    all_Price_list.append(Price.text)
all_Price_list
```

```python
Amazon_data = pd.DataFrame([all_Name_list,all_Rating_list,all_Review_list, all_Price_list])
```

```python
Amazon_data = Amazon_data.transpose()
Amazon_data.columns = ['Name','Rating','Review','Price']
```

```python
Amazon_data
```

```python
Amazon_data.to_excel('Amazon.xlsx', encoding='utf-8')
```
